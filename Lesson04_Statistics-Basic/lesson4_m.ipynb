```python
# NumPyをimport
import numpy as np
```

```python
# 魚の体長データ
fish_data = np.array([2,3,3,4,4,4,4,5,5,6])
```

```python
# 合計値
np.sum(fish_data)
```

```python
# 平均値
np.mean(fish_data)
```

```python
# 中央値
np.median(fish_data)
```

- ddof (Delta Degrees of Freedom) とは、分散や標準偏差を計算する際に、データ数から引く自由度調整の値を指します。特にPythonの`numpy.var()やnumpy.std()`などで使われ、ddof=0（デフォルト）は標本分散（不偏推定量ではない）、**ddof=1は不偏分散（標本から母集団の分散をより正確に推定するためのもの）**を計算します。これは、標本データから母集団の特性を推定する際に、分母をデータ数$N$ではなく$N-1$で割ることで、より正確な推定値（不偏推定量）を得るための調整です。
- ddofの役割   ddof=0 (デフォルト):\(N\)（データ数）で割る。「標本分散」を計算します。データそのもののばらつき（分散）を見たい場合に利用します。﻿ddof=1:\(N-1\)（データ数 - 1）で割る。「不偏分散」を計算します。データ全体（母集団）の分散を、手元の標本データから推定する際に、より偏りのない値（不偏推定量）を得るために使われます

- ddofは、**手元のデータが「標本」なのか「母集団全体」なのか、そして「データのばらつき」を知りたいのか(ddof=0)、「母集団の特性を推定したいのか(ddof=1)」**によって使い分けられる、分散・標準偏差計算における重要なパラメータ

```python
# 標本分散
np.var(fish_data, ddof=0)
```

```python
# 不偏分散
np.var(fish_data, ddof=1)
```

```python
# 標本標準偏差
np.std(fish_data, ddof=0)
```

```python
# 不偏標準偏差
np.std(fish_data, ddof=1)
```
