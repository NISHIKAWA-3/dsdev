```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("./sample_data/california_housing_train.csv")
```

```python
dataset.head()
```

```python
dataset1 = dataset[['population']]
dataset1.head()
```

```python
arr = dataset1.to_numpy()
print(arr)
```

```python
#合計値
np.sum(arr)
```

```python
#平均値
np.mean(arr)
```

```python
#中央値
np.median(arr)
```

```python
#標本分散
np.var(arr, ddof=0)
```

```python
#不偏分散
np.var(arr, ddof=1)
```

```python
#標本標準偏差
np.std(arr, ddof=0)
```

```python
#不偏標準偏差
np.std(arr, ddof=1)
```
